## Adatok előkészítése

```python
import numpy as np
```

```python
X_train = np.asarray(train_input[['customer_type','product_type_zart', 'sales_channel_dealer', 'onero', 'age_of_asset (honap)', 'Nace_low', 'Nace_medium', 'Nace_high']])
X_train[0:1]
```

```python
X_test = np.asarray(test_input[['customer_type','product_type_zart', 'sales_channel_dealer', 'onero', 'age_of_asset (honap)', 'Nace_low', 'Nace_medium', 'Nace_high']])
X_test[0:1]
```

```python
y_train = np.asarray(train_input['DPD90_12m'])
y_train [0:5]
```

## Modellezés

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
    
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20)
```

```python
yhat_test = model.predict(X_test)
yhat_test
```

```python
yhat_train = model.predict(X_train)
yhat_train
```

## Modell ereje

```python
from sklearn.metrics import roc_auc_score
roc=roc_auc_score(y_train, yhat_train)
gini=2*roc-1
gini
```
